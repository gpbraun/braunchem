```python
import matplotlib.pyplot as plt
from braunchem.chemistry.acidbase import WeakAcidSolution, StrongBaseSolution, Titration
```

```python
acid = WeakAcidSolution(Ka=[2e-5], C=4e-4, V=25)
base = StrongBaseSolution(C=1e-3, V=20)
```

```python
t = Titration(acid, base)
```

```python
x, y = t.curve()

plt.xlim([0, base.V])
plt.plot(x, y)
plt.show()
```
